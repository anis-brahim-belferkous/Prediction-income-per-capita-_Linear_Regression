```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("/kaggle/input/canada-income-per/Canada_income_per_capita.csv")
```

```python
df.info()
```

```python
sns.regplot(x="Year",y="Income",data=df)
```

```python
lri = LinearRegression()
```

```python
lri.fit(df[["Year"]].values,df.Income)
```

```python
lri.predict([[2021]])
```

```python
decimal_val_income = float(np.squeeze(lri.predict([[2021]])))
print(decimal_val_income)
```
